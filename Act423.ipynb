```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

data=pd.read_csv('GyC23.csv')
```

```python
data.info()
```

```python
data1=data.drop(["Unnamed: 0"],axis=1)
data1
```

```python
data2 =data1.fillna(method="ffill") 
```

```python
data2.info()
```

```python
#Filas cuantitativas
cuanti=data2.iloc[:,8:17]
```

```python
#Cualitativas
cuali=data2.iloc[:,[0,1,2,3,4,5,6,7,17,18,19]]
cualiC=cuali.fillna("N/A")
```

```python
cuanti.info()
```

```python
cualiC.info()
```

```python
fig=plt.figure(figsize=(15,8))
cuanti.plot(kind='box',vert=False)
plt.title("Valores atípicos del dataframe")
plt.show() #Dibujamos el diagrama
```

```python
y=cuanti
ls=y.mean() + 3*y.std()
li=y.mean() - 3*y.std()
print("Límite Superior Permitido", ls)
print("Límite Inferior Permitido", li)
```

```python
data4=cuanti[(y<=ls)&(y>=li)]
data4
```

```python
vn=data4.isnull().sum()
vn
```

```python
datac=data4.copy()
datac=datac.fillna(round(data4.mean(),1))
datac
```

```python
vn=datac.isnull().sum()
vn
```

```python
	
# Unimos la columna 'cuantitativas' con el dataframe
Datos_limpios = pd.concat([datac, cualiC], axis=1)
Datos_limpios
```

```python
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Datos_limpios23.csv")
```
